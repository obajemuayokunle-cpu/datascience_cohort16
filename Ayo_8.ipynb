Task 1 — Product Records Setup

```python
store_a_name = "Bliss store"
store_b_name = "Success secrets store"
store_a_prices =[12000,10000,15000,3400,25000,10000,15000]
store_b_prices = [13000.0,15000.0,35000.0,16000.0,17000.0,15000.0,13000.0]
```

Task 2 — Convert to Sets (Data Cleaning Stage)

```python
store_a_prices = set(store_a_prices)
print(store_a_prices)
```

```python
store_b_prices = set(store_b_prices)
print(store_b_prices)
```

Task 3 — Perform Set Operations

```python
common_prices = store_a_prices.intersection(store_b_prices)
print(Common_prices)
```

```python
only_a = store_a_prices.difference(store_b_prices)
print(only_a)
```

```python
only_b = store_b_prices.difference(store_a_prices)
print(only_b)
```

```python
all_unique_prices= store_a_prices.union(store_b_prices)
print(unique_prices)                                  
```

Task 4 — Update the Data (Set Methods Practice)

```python
store_a_prices.add(45000)
print(store_a_prices)
```

```python
store_b_prices.add(49000.0)
print(store_b_prices)
```

```python
store_a_prices.remove( 25000)
print(store_a_prices)
```

```python
store_b_prices.discard(49000.0)
print(store_b_prices)
```

Task 5 — Mix Data Types

```python
total_sale_a = sum(store_a_prices)
print(total_sale_a)
```

```python
total_sale_b = sum(store_b_prices)
print(total_sale_b)
```

```python
f"({store_a_name }) Total sales({total_sale_a})"
```

```python
f"({store_b_name})Total sales({total_sale_b})"
```

Task 6 — Final Report

```python
print("-----Final Report-----")
print(f"Store A Name: {store_a_name}")
print(f"Store B Name: {store_b_name}\n")

print("Store A Cleaned Prices:", store_a_prices)
print("Store B Cleaned Prices:", store_b_prices, "\n")

print("Common Prices:", common_prices)
print("Prices only in Store A:", only_a)
print("Prices only in Store B:", only_b)
print("All Unique Prices:", all_unique_prices, "\n")
```
