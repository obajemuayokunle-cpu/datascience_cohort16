```python
temperature = int(input("what is the temperature: "))
if temperature > 30:
    print ("Hot day")
else:
    print ("Normal day")

age = int(input("how old are you: "))
if age >= 18:
    print ("Eligible for registration")
else:
    print ("Not eligible")


product = ["shoes","clothes","bags","caps"]
if product in ["shoes"]:
    print ("Available")
else:
    print ("Not Available")

customer_id = [456,455,456,546,556,456,555,456]
unique_customer = set(customer_id)
print("Unique customer IDs: ", unique_customer)
print("Number of unique customers:", len(unique_customer))

rating = int(input("what is you rating: "))
if rating == 5:
    print ("excellent")
elif rating > 3:
    print ("needs improvement")
else:
    print ("good")

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

if num1 > num2:
    print("The first number is larger.")
elif num2 > num1:
    print("The second number is larger.")
else:
    print("Both numbers are equal.")

Email = {"ikly@gmail.com","gadrg@gmail.com","casds@gmail.com"}
user_email = input("what is your email: ")
if user_email in Email:
    print ("login succesful")
else:
    print ("email not found")

list_a = [24,34,32,12,34,16,17]
list_b = [12,13,43,23,12,18,23,45,65]

if len(list_a) > len(list_b):
    print ("List A is longer")
elif len(list_b) > len(list_a):
    print ("List B is longer") 
else:
    print ("Both lists have the same length")

```
